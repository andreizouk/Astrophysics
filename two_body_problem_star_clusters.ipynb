```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4*np.pi, 0.01)
y = np.sin(x)
y = y*np.exp(-x/4)

plt.figure(figsize=(16, 8))
plt.plot(x, y)

plt.title('Затухающие колебания маятника')
plt.ylabel('Координата, м') # у нас здоровый маятник
plt.xlabel('Время, с.')     # и видимо очень вязкая среда
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# 3,0856776⋅10+16 м = 30,8568 трлн км (петаметров) = 3,2616 светового года.
# m sum = 1,9885 · 10+30 kg
# G = 6,67430(15)⋅10−11 
parsec = 3.0856776*10**16

Mass_sun=1.9885*10**30

G = 6.674301*10**-11

#length = 1-200 parsec
#masses 1-10000 of sun mass

year= 31536000

Koeff = parsec*np.sqrt(parsec)/np.sqrt(Mass_sun)


M = 1 
l = np.arange(1, 200, 0.01)
y = np.pi*l*np.sqrt(l)/np.sqrt(2*G*M)*Koeff/year/2
#y1 = np.pi*l*parsec*np.sqrt(l*parsec)/np.sqrt(2*G*M*Mass_sun)#*Koeff
#y = y*np.exp(-x/4)

M = 100 
y1= np.pi*l*np.sqrt(l)/np.sqrt(2*G*M)*Koeff/year/2

M = 10000 
y2= np.pi*l*np.sqrt(l)/np.sqrt(2*G*M)*Koeff/year/2

S = 13.79*10**9*l/l

plt.figure(figsize=(16, 8))
plt.plot(l, y, l, y1, l , y2 , l , S)
plt.yscale('log')
plt.title('Время сближения скоплений / время жизни вселенной ')
plt.ylabel('Время сближения, лет') # у нас здоровый маятник
plt.xlabel('Начальное расстояние, парсек')     # и видимо очень вязкая среда
plt.show()

Koeff
```
